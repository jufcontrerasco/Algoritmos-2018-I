```python
from time import time
import matplotlib.pyplot as plt

limite = int(input("Introduzca el n-esimo numero de la serie de Fibonnaci que desea calcular: "))

tiempos = []
r = []
for i in range(0,limite):
    n1 = 0
    n2 = 1
    tiempo_inicial = time()

    if i>1:
        for k in range(0,i):
            n2 = n1+n2
            n1 = n2-n1
        print(n2)
    tiempo_final = time()

    if i==0:
        print (n1)
        tiempo_final = time()

    if i==1:
        print (n2)
        tiempo_final = time()

    if i<0:
        print('Error')
        tiempo_final = time()

    tiempo_ejecucion = tiempo_final-tiempo_inicial
    print('Tiempo de ejecucion: ',tiempo_ejecucion)
    tiempos.append(tiempo_ejecucion)
    r.append(i+1)

print(tiempos)
print(r)

plt.plot(r,tiempos)
plt.ylabel('Tiempo de ejecución')
plt.xlabel("N-esimo numero de la serie de Fibonacci")
plt.axis([0, limite, 0, max(tiempos)])
plt.show()
```
