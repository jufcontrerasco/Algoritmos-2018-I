```python
from time import time
import numpy as np
import matplotlib.pyplot as plt

f = int(input("Hasta que tama√±o de matriz nxn desea usar?: "))
tiempos = []
r = []
j = []

for k in range (0,10):
    for i in range (1, f+1):
        tiempo_inicio = time()
        a = np.full((i, i), 1)
        b = np.full((i, i), 2)
        np.matmul(a, b)
        tiempo_final = time()
        tiempo_ejecucion = tiempo_final-tiempo_inicio
        if k ==0:
            tiempos.append(tiempo_ejecucion)
        else:
            tiempos[i-1] = tiempos[i-1]+tiempo_ejecucion


for i in range(f):
    r.insert(0,tiempos[i]/10)
    j.append(i+1)

print(r)
plt.plot(j,r)
plt.ylabel('Tiempo de ejecuci√≥n promedio')
plt.xlabel("Tama√±o de matriz")
plt.axis([1, f, 0, max(r)])
plt.show()
```
